```python
import pandas as pd
data=pd.read_csv('Desktop/Datasets/spam.csv',encoding='latin-1')
data = data.drop(["Unnamed: 2", "Unnamed: 3", "Unnamed: 4"], axis=1)
data = data.rename(columns={"v1":"label", "v2":"text"})
data['label_num'] = data.label.map({'ham':0, 'spam':1})

#Split the dataset for training and test sets

from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(data["text"],data["label"], test_size = 0.2, random_state = 10)

#Make a matrix of 1-grams in the training set
#We can change it to 2-gram, 3-gram etc accordingly

from sklearn.feature_extraction.text import CountVectorizer
vect = CountVectorizer(ngram_range=(1, 1))

#Fit this to training set
vect.fit(X_train)


#Do the tfidf transformations in this stage

X_train_df = vect.transform(X_train)
X_test_df = vect.transform(X_test)


#Apply Multinomial Naive bayes algorithm to the dataset


from sklearn.naive_bayes import MultinomialNB
model = MultinomialNB()
model.fit(X_train_df,y_train)
prediction= model.predict(X_test_df)


#Calculate the accuracy


from sklearn.metrics import accuracy_score
accuracy_score(y_test,prediction)
```
